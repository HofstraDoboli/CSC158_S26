```python
import numpy as np

# Problem: S → A → G
# Reward at G = 1
# γ = 0.9
# Grid: 3 states in a line: S -> A -> G
# S=0, A=1, G=2

gamma = 0.9
alpha = 0.5

V = np.zeros(3)

# Reward only at goal
def reward(s):
    return 1 if s == 2 else 0

# Deterministic transition
def next_state(s):
    return min(s + 1, 2)

print("Initial V:", V)

# Simulate episodes
for episode in range(10):
    s = 0
    while s != 2:
        s_next = next_state(s)
        r = reward(s_next)

        # TD update
        V[s] = V[s] + alpha * (r + gamma * V[s_next] - V[s])

        s = s_next

    print(f"Episode {episode+1}: V = {V}")

print("\nFinal learned values:", V)
```
