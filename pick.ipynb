##select groups of students to present in class

```python
# open csv file
import pandas as pd
df_names = pd.read_csv('names.csv', sep = ';', names = ['name','email'])
print(df_names.head())
print(df_names.columns)
print(df_names.shape)
```

```python
# select a number of students randomly from the dataframe without replacement
import random
# remove the students that have already been selected in the previous step and stored in the new filter
prev_selected_students = pd.read_csv('selected_students.csv', sep=';', names=['name', 'email'], header=None)
print(prev_selected_students)
if len(prev_selected_students) > 0:
    print("Previous selected students:")
    print(prev_selected_students)
    filtered_df_names = df_names[~df_names['name'].isin(prev_selected_students['name'])]
else:
    print("No previously selected students found.")
    print(len(df_names))
    filtered_df_names = df_names
print(len(filtered_df_names))

random_students = filtered_df_names.sample(n=6, replace=False)
print(random_students['name'])
```

```python
# save the students to a new csv file or at append at the end of the file
random_students.to_csv('selected_students.csv', index=False, mode='a', header=not pd.io.common.file_exists('random_students.csv'))
```
